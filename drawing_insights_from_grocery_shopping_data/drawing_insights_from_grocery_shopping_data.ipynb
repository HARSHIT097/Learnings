Importing Libraries

```python
import pandas as pd
import numpy as np

%matplotlib inline
```

Dataset:

Columns/features

1. order_hour_of_day: The hour of the day when the order was placed (0 = 12am, 1 = 1am etc)
2. department: The department the products came from (e.g. alcohol, produce)
3. num_orders_hour: The number of products ordered from this particular department during this particular hour.
4. Tot_orders_dept: The total number of products ordered from this particular department across all hours of the day.


```python
train = pd.read_csv("InstacartOrdersByDepartment.csv")
train.head(10)
```

# Data analysis

```python
#At first we remove the unnecessary records of items 
#which are not in food department.
train.loc[~train['department'].isin(['babies','householdor','international','missing','other','personal care','pets','produce'])]  
```

```python
#The records of indiviual food departments are taked in separate varibles.

#Alcohol
alcohol = train[train['department'] == 'alcohol' ]
# Plotting alcohol records
fig = alcohol.plot(x='order_hour_of_day',y='num_orders_hour',title='PLotting',grid = True, color='r',figsize = (10,10))

#Bakery
bakery= train[train['department'] == 'bakery' ]
# Plotting bakery records
fig = bakery.plot(x='order_hour_of_day',y='num_orders_hour',title='PLotting',grid = True, color='g',figsize = (10,10),ax=fig)

#Beverages 
beverages= train[train['department'] == 'beverages' ]
#Plotting beverages records
fig = beverages.plot(x='order_hour_of_day',y='num_orders_hour',title='PLotting',grid = True, color='b',figsize = (10,10),ax=fig)

#Breakfast
breakfast= train[train['department'] == 'breakfast' ]
#Plotting breakfast records
fig = breakfast.plot(x='order_hour_of_day',y='num_orders_hour',title='PLotting',grid = True, color='k',figsize = (10,10),ax=fig)

#Bulk
bulk= train[train['department'] == 'bulk' ]
#Plotting bulk records
fig = bulk.plot(x='order_hour_of_day',y='num_orders_hour',title='PLotting',grid = True, color='y',figsize = (10,10),ax=fig)

#Canned goods
canned_goods= train[train['department'] == 'canned goods']
#Plotting canned goods
fig = canned_goods.plot(x='order_hour_of_day',y='num_orders_hour',title='PLotting',grid = True, color='c',figsize = (10,10),ax=fig)

#Dairy eggs
dairy_eggs= train[train['department'] == 'dairy eggs']
#Plotting dairy eggs
fig = dairy_eggs.plot(x='order_hour_of_day',y='num_orders_hour',title='PLotting',grid = True, color='m',figsize = (10,10),ax=fig)

#Deli
deli= train[train['department'] == 'deli']
#Plotting deli records
fig = deli.plot(x='order_hour_of_day',y='num_orders_hour',title='PLotting',grid = True, color='r',marker='o',figsize = (10,10),ax=fig)

#Frozen
frozen= train[train['department'] == 'frozen']
#Plotting frozen records
fig = frozen.plot(x='order_hour_of_day',y='num_orders_hour',title='PLotting',grid = True, color='g',marker='o',figsize = (10,10),ax=fig)

#Dry goods and pasta
dry_goods_pasta = train[train['department'] == 'dry goods pasta']
#Plottin records for dry goods and pasta
fig = dry_goods_pasta.plot(x='order_hour_of_day',y='num_orders_hour',title='PLotting',grid = True, color='b',marker='o',figsize = (10,10),ax=fig)

#Meat and seafood
meat_seafood = train[train['department'] == 'meat seafood']
#Plotting meat seafood records
fig = meat_seafood.plot(x='order_hour_of_day',y='num_orders_hour',title='PLotting',grid = True, color='c',marker='o',figsize = (10,10),ax=fig)

#Pantry
pantry= train[train['department'] == 'pantry']
#Plotting pantry records
fig = pantry.plot(x='order_hour_of_day',y='num_orders_hour',title='PLotting',grid = True, color='y',marker='o',figsize = (10,10),ax=fig)

#Snacks
snacks= train[train['department'] == 'snacks']
#Plotting snacks records
fig = snacks.plot(x='order_hour_of_day',y='num_orders_hour',title='PLotting',grid = True, color='m',marker='o',figsize = (15,10),ax=fig)

fig.legend(['alcohol','bakery','beverages','breakfast','bulk','canned goods','dairy eggs','deli','frozen','dry goods pasta','meat seafood','pantry','snacks'])
fig.set_xlabel('Order hour of day')
fig.set_ylabel('Num orders hour')
fig.set_xlim([0,23])
```

```python
fig = breakfast.plot(x='order_hour_of_day',y='num_orders_hour',title='PLotting',grid = True, color='r',figsize = (10,10))

fig = meat_seafood.plot(x='order_hour_of_day',y='num_orders_hour',title='Breakfast vs meat seafoods',grid = True, color='c',figsize = (10,10),ax=fig)


fig.legend(['breakfast','meat seafood'])
fig.set_xlabel('Order hour of day')
fig.set_ylabel('Num orders hour')
fig.set_xlim([0,23])
```

Explanation of the results/Insights:
The result make sense before
in the morning people are more likely to buy breakfast than meat-seafoods and
in the afternoon people are less likely to buy breakfast and more likely to buy meat and
seafoods for supper or dinner.

Take away message:
It is found that there is more demand for breakfast than meat-seafoods in the morning
so keeping a balanced supply at different times is important for these two food departments

Limitations:
1. Data does not tell anything about demand for a particular food item,
it shows the relation of orders based on different departments
