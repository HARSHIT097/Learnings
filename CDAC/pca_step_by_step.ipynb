```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
```

```python
iris_df = pd.read_csv("J:\\cdac gdrive material\\ML\\Practical Machine Learning\\Datasets\\iris.csv")
iris_df.shape
```

```python
iris_df.head()
```

```python
# selecting only the features
X = iris_df.iloc[:, :-1]
```

```python
# scaling the data

scaled_X = StandardScaler().fit_transform(X)
scaled_X
```

```python
# Covariance matrix

features = scaled_X.T
cov_mtrx = np.cov(features)
cov_mtrx
```

```python
# Eigen decomposition

eig_values, eig_vectors = np.linalg.eig(cov_mtrx)
```

```python
eig_values
```

```python
eig_vectors
```

```python
# variance explained by each component

exp_var = []   # for storing the variance explained by each component

for i in range(len(eig_values)):
    exp_var.append(eig_values[i] / np.sum(eig_values))

exp_var
```

```python
# components 
principal_comp = scaled_X.dot(eig_vectors.T[0])
comp_2 = scaled_X.dot(eig_vectors.T[1])
```

```python
pca_iris = pd.DataFrame(principal_comp, columns=['PC!'])
pca_iris['PC2'] = comp_2
pca_iris
```
