# Question 2 – Minimum Swaps to Symmetric String

Question:

Given a lowercase alphabet string s, find the minimum number of swaps required to make it a
symmetry string. If it's not possible, then return -1.

Constraints:
n ≤ 1,000 where n is length of s.

Example 1
Input
s = "aabb"

Output
2

Explanation
 We can swap the middle "a" and "b" and then swap the first two "a" and "b" to get

"baab".

Example 2

Input
s = "aab"

Output
1
Explanation
We can swap the middle "a" and "b" to get "aba".

Example 3

Input
s = "abca"

Output
-1

Explanation
No Possibilities.

```python
def CountSwap(input_str, len_str):
    #coversion to list for indexing
    s = list(input_str)
    n = len_str
    print("length of string",len_str)
    print(s)
    # Intializing Counter to count minimum swap
    count = 0
    ans = True
 
    # A loop which run in half string
    # from starting
    for i in range(n // 2):
 
        # Left pointer
        left = i
 
        # Right pointer
        right = n - left - 1
 
        # A loop which run from right pointer
        # to left pointer
        while left < right:
 
            # if both char same then
            # break the loop if not
            # same then we have to move
            # right pointer to one step left
            if s[left] == s[right]:
                print("String found equal",count)
                break
            else:
                right -= 1
 
        # it denotes both pointer at
        # same position and we don't
        # have sufficient char to make
        # palindrome string
        if left == right:
            ans = False
            print("symmetry found total count", count)
            break
        else:
            for j in range(right, n - left - 1):
                (s[j], s[j + 1]) = (s[j + 1], s[j])
                print(count)
                print("")
                count += 1
    if ans:
        return (count)
    else:
        print("No possible swaps")
        return -1

if __name__ == '__main__':
  input_str = input("Enter the sentence:")
  input_str = input_str.lower()
  print("Entered string for evaluation:", input_str)
  #input_str = "aabbaa"
  len_str = len(input_str)
  print("####################")
  
  CountSwap(input_str, len_str)
    # Function calling
  ans1 = CountSwap(input_str, len_str)
  print("forward",ans1)
  print("####################")
  ans2 = CountSwap(input_str[::-1], len_str)
  print("backward",ans2)
  print("####################")
  print("Possible number of swaps:",max(ans1, ans2))
  print("####################")
  print(" -1 indicates no possible swaps")
  print("####################")
```
